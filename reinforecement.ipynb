```python
def sumli(li,a=0):
    sum=0
    for i in range(len(li)):
        if a > i:
              continue
        sum+=li[i]
    return sum
```

```python
li=[-1,-1,10]
sumli(li,2)
```

```python
score = [0 for i in range(env.observation_space.n)]
for itera in range(5):
    env.reset()
    env.env.s=0
    #env.render()
    visited_states=[]
    visited_states_reward=[]
    done=False
    while not done:
        oldstate = env.env.s
        action=policy[oldstate]

        state, reward, done, info = env.step(action)
        if (reward==0 and done== True):
            reward=-1
        elif reward == 0:
            reward=-0.04
        visited_states.append(oldstate)
        visited_states_reward.append(reward)
      #  print("s="+str(oldstate),"a"+str(action),"s'="+str(state),"reward="+str(reward))
        #env.render()
    v_states=[]
   # print(visited_states)
   # print(visited_states_reward)
    for i in range (len(visited_states)):
        if visited_states[i] not in v_states:
            v_states.append(visited_states[i])
            if score[visited_states[i]] !=0:
            #print("state"+ str(visited_states[i])+"="+str(sumli(visited_states_reward,i)))
                score[visited_states[i]]+=sumli(visited_states_reward,i)
                score[visited_states[i]]/=2
            else:
                score[visited_states[i]]=sumli(visited_states,i)
        print(score)
```
